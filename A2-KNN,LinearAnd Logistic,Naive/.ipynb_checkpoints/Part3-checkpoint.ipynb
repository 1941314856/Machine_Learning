```python
split = int(0.8 * df.shape[0])
training_data = df[:split] # 80% of the total data
testing_data = df[split:]  # 20% of the total data
```

### Feature Normalisation
${x_i}$= $\frac{x_i - \mu}{\sigma}$ 

Feature Normalisation is done because data in some columns is very small in comparison to other columns data.

```python
columns=training_data.shape[1]
#print(training_data.shape)
X=training_data.iloc[:,0:columns-1]# features Sets
mu=X.mean()
sigma=X.std()

# features normalisation
X=(X-X.mean())/X.std()

Y=training_data.iloc[:,columns-1:columns] # outputSet
X.insert(0, 'Ones', 1)
print(Y.shape)
print(X.shape)
X.head()
```

> Convert the X and Y into numpy matrix because we are going to do the vectorised Implementation

```python
X = np.matrix(X.values)
Y = np.matrix(Y.values)
# print(X.shape)
# print(Y.shape)

```

#### Mean Square Error
${J(\theta)}$=${\frac{1}{2*m}}$${\sum_{i=0}^{m}}$(${\hat{y_i}-{y_i}})^2$


J: is cost function


m : no. of training examples

${\theta}$: parameters

```python
def costMeanSquare(X,Y,theta):
    j=0.0
    predictions=np.power((np.dot(X,theta.T)-Y),2)
    j=np.sum(predictions)
    return j/(2*len(X))
```

#### Mean Absolute Error
${J(\theta)}$=${\frac{1}{m}}$${\sum_{i=0}^{m}}$${\left|{\hat{y_i}}-{y_i}\right|}$


J: is cost function


m : no. of training examples

${\theta}$: parameters

```python
def costMeanAbsolute(X,Y,theta):
    j=0.0
    predictions=abs(np.dot(X,theta.T)-Y)
    j=np.sum(predictions)
    return j/(len(X))
```

#### Mean Absolute Percentage Error
${J(\theta)}$=${\frac{100%}{m}}$${\sum_{i=0}^{m}}$${\left|\frac{{\hat{y_i}}-{y_i}}{y_i}\right|}$


J: is cost function


m : no. of training examples

${\theta}$: parameters

```python
def costAbsolute_Per_Error(X,Y,theta):
    j=0.0
    predictions=(np.dot(X,theta.T)-Y)/Y
    j=np.sum(abs(predictions))
    return (100*j)/len(X)
    
```

#### Gradient Descent algo
repeat Untill Converges{


${{\theta_j} :=}$${{\theta_j}}$-${\alpha}$*${\frac{\partial}{\partial {\theta_j}} J(\theta)}$


}

${\alpha}$: Learning rate constant

```python
#Vectorised Implementation
def gradientDescent(X, y, theta, alpha, iters):
    
    Jhistory=np.zeros(iters)
    JmeanAbs=np.zeros(iters)
    JmeanAbsPerErr=np.zeros(iters)
    ThetaforCoffiecients=np.zeros((iters,theta.shape[1]))
    m=len(X)
    for i in range(iters):
        pre = (X*theta.T)-y
        #print(pre.shape)
        delta=np.dot(np.transpose(pre),X)
        theta=theta-(alpha/m)*delta
        
        ThetaforCoffiecients[i]=theta
        Jhistory[i] = costMeanSquare(X, y, theta)
        JmeanAbs[i]=costMeanAbsolute(X,y,theta)
        JmeanAbsPerErr[i]=costAbsolute_Per_Error(X,y,theta)
        
    return theta,Jhistory,JmeanAbs,JmeanAbsPerErr,ThetaforCoffiecients
    
```

#### Behavoiur of ${ \alpha} $ 
1. If we put ${\alpha=}$ some -ve value like ${-.001}$ then we can see that cost is increasing w.r.t to no of iterations.
2. Generally , we take alpha value initially some like .001 , then kept increasing by multiple of 3 and keep seeing the behavour of cost w.r.t to ${\alpha}$ and no of iterations. 
3. ${\alpha}$ can't be too big because it overshoot the gradient and can't be too small because it will be too slow .
4. Change the alpha , see the change in behaviour of cost function.

```python
alpha=.009
iters=1000
theta = np.matrix(np.array([0,0,0,0,0,0,0,0]))
minTheta, cost2,cost1 ,cost0,Big= gradientDescent(X, Y, theta, alpha, iters)
print(minTheta)
```

```python
ig, ax = plt.subplots(figsize=(10,7))
ax.plot(np.arange(iters), cost2, 'r')
ax.set_xlabel('Iterations')
ax.set_ylabel('Cost')
ax.set_title('Visualisation of change in Cost w.r.t to iterations')
```

```python
ig, ax = plt.subplots(figsize=(10,7))
ax.plot(np.arange(iters), cost1, 'o')
ax.set_xlabel('Iterations')
ax.set_ylabel('Cost')
ax.set_title('Visualisation of change in MeanAbsolute w.r.t to iterations')
```

```python
ig, ax = plt.subplots(figsize=(10,7))
ax.plot(np.arange(iters), cost0, 'orange')
ax.set_xlabel('Iterations')
ax.set_ylabel('Cost')
ax.set_title('Visualisation of change in costAbsolute_Per_Error w.r.t to iterations')
```

```python
columns=testing_data.shape[1]
X_test=testing_data.iloc[:,0:columns-1]# features Sets
Y_test=testing_data.iloc[:,columns-1:columns] # outputSet

print(X_test.shape)
print(Y_test.shape)

```

```python
def prediction(test_data):
    a=np.matrix(np.array([1]))
    m=np.matrix(np.array(mu))
    s=np.matrix(np.array(sigma))
    test=np.matrix(np.array(test_data))
   
    test=(test-m)/s
    #print(test)
    
    test=np.hstack((a,test))
    
    return np.dot(test,minTheta.T)
        
```

```python
testSet = [315,103,2,2.5,2,8.5,0]# 7 features values

result = prediction(testSet)
print(result)
```

```python
from sklearn.linear_model import LinearRegression

a=np.matrix(np.array([1]))
m=np.matrix(np.array(mu))
s=np.matrix(np.array(sigma))
test=np.matrix(np.array(testSet))
test=(test-m)/s
test=np.hstack((a,test))

model = LinearRegression()
model.fit(X, Y)

X_predict = np.matrix(np.array(test)) 
y_predict = model.predict(X_predict)
print(y_predict)
```

```python
output=[]#empty list
for i in range(0,len(X_test)):
    a=prediction(X_test.iloc[i].tolist())
    b=np.asscalar(a)
    output.append(b)

#print(output)
```

```python
from sklearn.metrics import r2_score
r2_score(list(Y_test.values), output)
```
