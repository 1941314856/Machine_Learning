### Algorithms for Housing Prediction using  Neural Networks with 3 Hidden Layers

###  Algorithm

1.    Define the Neural Network Structure ( no. of nodes in each hidden layers,  learning rate etc) 
2.    Initialize the model's parameters.
3.    Loop:
        - Implement forward propagation
        - Compute loss
        - Implement backward propagation to get the gradients
        - Update parameters (gradient descent)
        
        
 4.   Combine all functions
 5.    Prediction

#### Data Preprocessing: Before Anything 
> 1. You must the Data Preprocessing like some column have Nan or zero or may missing value.<br/>
> 2. Either fill with zero or mean , it varies from data to data.<br/>
> 3. Try to remove those features which are redundant because those features can be learn in hidden layers.
> 4. Encode all the strings input to the categorical , so that it cause no problem.
> 5. To improve Performance we can do  data preparation techniques like Standardization/Normalization.


1. **Neural network structure**

${\bullet}$  In  this, No of perceptron in Input layers = No of features i.e 80(we drop id column)

${\bullet}$  In this , Output is not a vector but a real no. i.e price of house , so we have to use only 1 perceptron in output layer

2. **Initialisation** : Initialise the weights and other parameters like in previous question.

3. **Activation Function**:
> 1.  ${In   Hidden   Layer}$: Suggestion is to use the ${ReLU}$  in hidden layers because it works better than ${tanh}$ and ${sigmoid}$.<br/>
>2.  ${Output Layer}$: We have to do the prediction of Housing price i.e a real no , so use of ${ReLU}$ is must in output Layer or you can use the ${Leaky ReLU}$ or Simply use the Linear Activation Function.

4. **Cost Function:**
> We have to use the Mean Square Error Function instead of cross entropy because this is a Supervised Regression Task.

 ${J(\theta) = \frac{1}{2*m}\Sigma_{i=1}^{m}{(y_i-h(\theta)_i})^2}$

5. **Optimization**:
>  You can use the BackPropogation and try to Model with different hyperparameters to use the optimise the cost.
